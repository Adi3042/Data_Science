```python
Q1. What is the mathematical formula for a linear SVM?


    
    The mathematical formula for a linear SVM is:-

        w^Tx + b = 0
        
    where:-
        w is a weight vector
        x is an input vector
        b is a bias term
```

```python
Q2. What is the objective function of a linear SVM?


    
    (i). The objective function of a linear SVM is to maximize the margin between the two classes, while minimizing the regularization penalty.
    (ii). The margin is defined as the distance between the hyperplane and the closest data points from each class.
    (iii). The regularization penalty is used to prevent the model from overfitting the training data.
    
    Objective f^n:-

        Minimize: 1/2 * ||w||^2
        
    Constraints:-

        y_i * (w^T x_i + b) >= 1 for all i
        
        where:-
            w is the weight vector
            b is the bias term
            x_i is the i-th training sample
            y_i is the i-th target label
```

```python
Q3. What is the kernel trick in SVM?


    
    (i). The kernel trick in SVM is a technique that allows us to learn non-linear decision boundaries using a linear SVM.
    
    (ii). It does this by transforming the data into a higher-dimensional space, where it is more likely to be linearly separable.
    
    (iii). we simply need to define a kernel function, which is a function that takes two data points as input and returns a similarity score. 
    
    (iv). The transformed data is obtained by applying the kernel function to each pair of data points.
    
    (v). In short, the kernel trick in SVM is a way to make linear SVMs more powerful by transforming the data into a higher-dimensional space.
```

```python
Support vectors in SVM play a critical role in defining the decision boundary and maximizing the margin between the two classes. 
They are the data points that are closest to the hyperplane, and they are the only data points that contribute to the objective function of the SVM.

In the above example, the data points are colored red and blue, depending on the class to which they belong. 
The hyperplane is drawn in black, and it separates the red and blue data points. 
The support vectors are the data points that are closest to the hyperplane.

If we remove any of the support vectors, the hyperplane will move, and the margin between the two classes will decrease. 
This is why support vectors are so important in SVM.

Some of the advantages of using support vectors in SVM:-

    Support vectors define the decision boundary and maximize the margin between the two classes.
    
    Support vectors are the only data points that contribute to the objective function of the SVM.
    
    Support vectors can be used to represent the entire decision boundary of the SVM.
    
    SVM is a powerful machine learning algorithm for classification and regression tasks. 
    
    Support vectors play a critical role in making SVM so effective.
```

```python
Q5. Illustrate with examples and graphs of Hyperplane, Marginal plane, Soft margin and Hard margin in SVM?


Hyperplane:-

    In a linear SVM, the hyperplane is the decision boundary that separates two classes.
    It is represented by the equation---
    
                wâ‹…x+b=0, where 

        w is the weight vector and b is the bias term.
        
    The hyperplanes equation depends on the choice of w and b, and it maximizes the margin between classes.

Marginal Plane:-

    The marginal planes are parallel to the hyperplane and are used to define the margin. 
    
    In a hard-margin SVM, these planes are positioned to be equidistant from the hyperplane, touching the closest data points. 
    
    In a soft-margin SVM, the marginal planes can tolerate some data points falling within the margin.

Hard Margin:-

    In a hard-margin SVM, the margin is maximized, and there is no tolerance for data points within the margin.
    
    It assumes that the data is linearly separable, and any data point inside the margin or on the wrong side of the hyperplane is considered an error.
    
    Hard-margin SVMs are sensitive to outliers.

Soft Margin:-

    In a soft-margin SVM, some tolerance is allowed for data points to fall within the margin or on the wrong side of the hyperplane.
    
    It is used when data is not perfectly separable or when the goal is to achieve better generalization.
    
    Soft-margin SVMs introduce a parameter C that controls the trade-off between maximizing the margin and minimizing classification errors.
```
