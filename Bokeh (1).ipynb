```python
You need to install the Bokeh library first. You can install it using pip:
    
            pip install bokeh
        
Import the necessary Bokeh modules and functions:
    from bokeh.plotting import figure, show, output_file
    from bokeh.io import output_notebook
```

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, show, output_file
from bokeh.sampledata.iris import flowers
```

```python
output_file("temp.html")

p = figure(title = "Flower")
p.xaxis.axis_label = "Patel Length"
p.yaxis.axis_label = "Patel Width"
p.circle(flowers["petal_length"], flowers["petal_width"])
show(p)
```

```python
Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
```

```python
In Bokeh, glyphs are visual elements that represent data on a plot. 
They can be used to display points, lines, bars, patches, and other geometric shapes. 
Glyphs are added to a Bokeh plot to visualize and represent data.
```

```python
'''
    Here's a complete example that demonstrates adding of circle glyphs to a Bokeh plot:---
'''
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_file("temp.html")

p = figure(title = "Bokeh Plot")

p.circle([1, 2, 3, 4, 5], [6, 7, 2, 4, 5])

p.xaxis.axis_label = "X-axis"
p.yaxis.axis_label = "Y-axis"
show(p)
```

```python
Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?
```

```python
'''
    To customize the appearance of a Bokeh plot, including the axes, title, and legend, you can modify various 
    properties of the plot object. Here are some common customizations you can apply:----
'''
output_file("temp.html")

p = figure(title = "Bokeh Plot")

p.scatter([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], fill_color = "red", legend_label = "HIGH", size = 20)

p.xaxis.axis_label = "Customized X-axis"
p.yaxis.axis_label = "Customized Y-axis"

show(p)
```

```python
    A Bokeh server is a Python library that allows you to create and deploy interactive web applications 
    and dashboards with Bokeh plots. It enables you to create dynamic and responsive plots that can be 
    updated in real time based on user interactions or streaming data.
    
    Here's how you can use the Bokeh server to create interactive plots that can be updated in real time:---

    (i) Create the initial Bokeh plot using the Bokeh library.
    (ii) Write a Python script that defines the Bokeh server application.
    (iii) Configure callbacks in the application to handle user interactions and updates.
    (iv) Run the Bokeh server by executing the script.
    (v) Open a web browser and navigate to the provided URL.
    (vi) Interact with the plot and see it update in real time based on callbacks and user input.
    (vii) The Bokeh server enables you to build dynamic, data-driven visualizations that respond to user actions and can be updated in real time.
```
