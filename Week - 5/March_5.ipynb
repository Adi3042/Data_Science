#### **Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.** ####

In Bokeh, glyphs are visual elements that represent data on a plot. They can be used to display points, lines, bars, patches, and other geometric shapes. Glyphs are added to a Bokeh plot to visualize and represent data.

To add glyphs to a Bokeh plot, you typically use the GlyphRenderer class and its associated glyph methods provided by Bokeh.

Here's an example of how to add glyphs to a Bokeh plot:

1. Create a new Bokeh plot using the `figure()` function.

2. Use the `circle()` method to add a circle glyph to the plot. Specify the x and y coordinates of the data points, size, and color.

3. Use the `line()` method to add a line glyph, specifying the x and y coordinates, line width, and color.

4. Use the `square()` method to add a square glyph, specifying the x and y coordinates, size, fill color, and line color.

Finally, display the plot using the `show()` function.

This example demonstrates how to add different types of glyphs (circle, line, square) to a Bokeh plot, each with its own set of attributes such as size, color, and line width.


````python
## Example that demonstrates circle uses
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_file("temp.html")

p = figure(title = "Bokeh Plot")

p.circle([1, 2, 3, 4, 5], [6, 7, 2, 4, 5])

p.xaxis.axis_label = "X-axis"
p.yaxis.axis_label = "Y-axis"
show(p)
````

#### **Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?** ####

To customize the appearance of a Bokeh plot, including the axes, title, and legend, you can modify various properties of the plot object. Here are some common customizations you can apply:

````python
output_file("temp1.html")

p = figure(title = "Bokeh Plot")

p.scatter([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], fill_color = "red", legend_label = "HIGH", size = 20)

p.xaxis.axis_label = "Customized X-axis"
p.yaxis.axis_label = "Customized Y-axis"

show(p)
````

#### **Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time?** ####

A Bokeh server is a Python library that enables the creation and deployment of interactive web applications and dashboards featuring Bokeh plots. It empowers you to generate dynamic and responsive plots that can be updated in real time, responding to user interactions or streaming data.

Here's how you can utilize the Bokeh server to create interactive plots that can be updated in real time:

1. Create the initial Bokeh plot using the Bokeh library.

2. Write a Python script that defines the Bokeh server application.

3. Configure callbacks in the application to handle user interactions and updates.

4. Run the Bokeh server by executing the script.

5. Open a web browser and navigate to the provided URL.

6. Interact with the plot and observe real-time updates based on callbacks and user input.

The Bokeh server provides the capability to construct dynamic, data-driven visualizations that respond to user actions and can be updated in real time.

#### **Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?** ####

To embed a Bokeh plot into a web page or dashboard using Flask or Django, follow these steps:

1. Create the Bokeh plot using the Bokeh library.

2. Save the plot to an HTML file using Bokeh's `save()` function.

3. Integrate Flask or Django into your web application.

4. Define a route in your Flask or Django application to handle the plot's display.

5. Create an HTML template that will render the page with the embedded plot.

6. Load the HTML file containing the Bokeh plot in the route handler function.

7. Render the HTML template and include the Bokeh plot within it.

8. Start your Flask or Django application and access the URL to view the embedded Bokeh plot.
