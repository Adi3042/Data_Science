### **Q3- What is the difference between AI, ML, DL, and DS?** ###

<img src="https://www.deviq.io/hubfs/DevIQ%20website%20active%20images/ai-ml-dl-ds-venn-diagram-deviq.png" alt="AI, ML, DL, and DS Venn Diagram">

**Artificial Intelligence (AI):** AI refers to the broader concept of machines or systems that can perform tasks that typically require human intelligence. AI aims to create systems that can reason, learn, perceive, understand natural language, and interact with the environment.

**Machine Learning (ML):** ML is a subset of AI that focuses on the development of algorithms and models that enable computers to learn from data and make predictions or decisions.

**Deep Learning (DL):** DL is a specialized subset of machine learning that specifically deals with neural networks containing multiple layers.

**Data Science (DS):** DS involves the application of various techniques, including statistics, ML, and AI, to extract insights and knowledge from data.

### **Q5- What are the main differences between Supervised, unsupervised, and semi-supervised learning?** ###

**Supervised Learning** deals with labeled data and aims to make accurate predictions or classifications based on the provided labels.

**Unsupervised Learning** focuses on discovering patterns and relationships within unlabeled data without explicit target values.

**Semi-Supervised Learning** combines labeled and unlabeled data to improve model performance and leverage the benefits of both supervised and unsupervised approaches.

---
### **Q6- What is train, test, and validation split? Explain the importance of each term.** ###

**Training Set:** This is where the model learns from the data, capturing patterns and relationships. A well-structured training set ensures the model's ability to generalize and make accurate predictions.

**Validation Set:** Fine-tuning the model using the validation set helps optimize hyperparameters and prevent overfitting, ensuring that the model performs well on new data beyond the training set.

**Test Set:** The test set provides an unbiased assessment of the model's real-world performance, verifying its ability to generalize and make reliable predictions on unseen data.

### **Q7- How can unsupervised learning be used in anomaly detection?** ###

Anomaly detection involves identifying data points that deviate significantly from the expected or normal behavior.

Unsupervised learning can be effectively used in anomaly detection by leveraging its ability to identify patterns and outliers within data without requiring labels.

**Examples of anomalies:**
- Isolation Forest
- Autoencoders
- Statistical Approaches
