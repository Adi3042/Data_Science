### **Q1. What are the three measures of central tendency?**

The three measures of central tendency are:-
1. Mean
2. Median
3. Mode

---

### **Q2. What is the difference between the mean, median, and mode? How are they used to measure the central tendency of a dataset?**

The mean is the average of all the values in a dataset. 

The median is the middle value when the data is arranged in order. 

The mode is the value that appears most frequently in the dataset. 

They are used to describe where the "center" of the data is located and provide insight into the typical or most representative value in the dataset.

### **Q5. How are measures of dispersion such as range, variance, and standard deviation used to describe the spread of a dataset?**

**Range:** The difference between the maximum and minimum values.

**Variance:** The average of the squared differences from the mean.

**Standard Deviation:** The square root of the variance.

**Example:**
Consider two datasets -

**Dataset A:**
[5, 10, 15, 20, 25]

**Dataset B:**
[5, 5, 15, 25, 25]

Both datasets have the same mean (15), but Dataset A has a larger standard deviation (8.66) compared to Dataset B (10).

This indicates that the values in Dataset A are more spread out than in Dataset B.

### **Q8. What do you understand about skewness in data?**

Skewness in data refers to the asymmetry in the distribution of data points. 

It indicates whether the data is skewed to one side or the other relative to the mean.

Positive skewness means the data is skewed to the right, and negative skewness means the data is skewed to the left.

---

### **Q9. If a data is right skewed then what will be the position of median with respect to mean?**

In a right-skewed distribution, the median will be less than the mean. This is because the longer tail of the distribution pulls the mean in that direction, making it higher than the median.

### **Q10. Explain the difference between covariance and correlation. How are these measures used in statistical analysis?**

Covariance measures the degree to which two variables change together. It indicates the direction of the linear relationship between variables but does not provide a standardized measure.

Correlation is a standardized measure that indicates both the strength and direction of the linear relationship between two variables.

Covariance and correlation are used in statistical analysis to understand how variables are related and to make predictions or decisions based on their relationships.

### **Q11. What is the formula for calculating the sample mean? Provide an example calculation for a dataset.**

**Sample mean (x̄):**

    x̄ = (sum of all data points) / (number of data points)

**Example:**

    For a dataset [10, 15, 20, 25, 30], the sample mean:
    x̄ = (10 + 15 + 20 + 25 + 30) / 5
    x̄ = 20

### **Q12. For a normal distribution data what is the relationship between its measure of central tendency?**

For a normal distribution, the mean, median, and mode are all located at the same point, which is the center of the distribution.

### **Q13. How is covariance different from correlation?**

Covariance measures the extent to which two variables change together, while correlation measures the strength and direction of the linear relationship between two variables, standardized to a range between -1 and 1.

### **Q14. How do outliers affect measures of central tendency and dispersion? Provide an example.**

Outliers can significantly affect measures of central tendency and dispersion:

**Mean:**
Outliers can pull the mean away from the center of the data.

**Median:**
Outliers have less impact on the median since it is less sensitive to extreme values.

**Standard Deviation:**
Outliers can increase the standard deviation, indicating higher variability.

**Example:**

In a dataset [10, 15, 20, 25, 100], the presence of the outlier 100 will increase the mean, but the median will remain relatively unaffected, leading to a higher standard deviation.
