**Q1. How does bagging reduce overfitting in decision trees?**

(i). Bagging reduces overfitting in decision trees by creating multiple bootstrap samples and training individual decision trees on each sample.
  
(ii). These individual trees tend to capture different aspects of the data due to the randomness introduced by bootstrapping.

(iii). When their predictions are combined, the ensemble model's variance decreases.

(iv). This reduction in variance helps reduce overfitting because the ensemble model is less sensitive to noise and specific patterns in the training data, resulting in a more generalized and robust model.

---

**Q2. What are the advantages and disadvantages of using different types of base learners in bagging?**

**Advantages of using different types of base learners in bagging:**

(i). Increased diversity in predictions, which can lead to better ensemble performance.
(ii). Effective for capturing different relationships in the data.
(iii). Can reduce the risk of overfitting.

**Disadvantages:-**

(i). If base models are too different, it may be challenging to combine their predictions effectively.
(ii). Managing and tuning a mix of diverse base learners can be more complex.

---

**Q3. How does the choice of base learner affect the bias-variance tradeoff in bagging?**

The choice of base learner in bagging can affect the bias-variance tradeoff:- 
Typically, bagging reduces the variance of the base learner, making it less prone to overfitting. 

(i). This reduction in variance can lead to a more balanced bias-variance tradeoff by reducing the model's tendency to fit the noise in the training data. 

(i).  However, the bias of the base learner is usually preserved or slightly increased in the ensemble. 

(i). In summary, bagging tends to decrease the variance of the base learner while potentially increasing its bias slightly, resulting in a more stable and less overfit model.

---

**Q4. Can bagging be used for both classification and regression tasks? How does it differ in each case?**

**Classification:-**

(i). In classification tasks, bagging typically involves training multiple base classifiers on bootstrapped samples of the training data and combining their predictions by majority voting. 
        
(ii). It helps improve the accuracy and robustness of the classification model, especially when dealing with noisy or imbalanced datasets.

**Regression:-**

(i). In regression tasks, bagging involves training multiple base regression models on bootstrapped samples of the training data and aggregating their predictions by averaging. The result is a more stable and less sensitive regression model, which is less likely to overfit and provides more reliable predictions.

The key difference lies in the aggregation method: majority voting for classification and averaging for regression.

---

**Q5. What is the role of ensemble size in bagging? How many models should be included in the ensemble?**

The ensemble size in bagging refers to the number of base models that are trained on different bootstrap samples and combined to form the final ensemble model.
    
The choice of ensemble size is a hyperparameter that can affect the performance of the bagging ensemble.

**Larger Ensemble Size:-**
    
Increasing the number of base models in the ensemble tends to improve the ensembles' stability and reduce variance further.
        
However, there may be diminishing returns, and it can increase computational complexity.

**Smaller Ensemble Size:-**
    
Using too few base models may not harness the full potential of bagging and may result in less improvement over a single model.

The optimal ensemble size often depends on the specific problem, dataset, and computational resources available. 
Cross-validation or grid search can help determine the best ensemble size for a given task.
