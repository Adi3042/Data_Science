```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,200)
y = np.sin(x)
plt.plot(x,y)
```

```python
x = np.linspace(0,10,200)
y = np.sin(x)
plt.plot(x,y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Sin Graph")
plt.show()
```
