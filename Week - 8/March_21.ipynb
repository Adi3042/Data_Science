**Q1. What is the difference between Ordinal Encoding and Label Encoding? Provide an example of when you might choose one over the other.**

**Ordinal Encoding:**
This method is used when there is a natural order or hierarchy among the categories in a categorical variable. Each category is assigned a unique integer based on its position in the order. *For example,* "Low," "Medium," and "High" could be encoded as 0, 1, and 2, respectively.

**Label Encoding:**
Label Encoding is used when there is no inherent order among the categories. It assigns a unique integer to each category without considering any order. *For example,* if you have categories like "Red," "Green," and "Blue," they might be encoded as 0, 1, and 2, respectively.

Use Ordinal Encoding when there is a clear order among categories, and that order is meaningful in your analysis. For example, for customer satisfaction levels (Low, Medium, High), ordinal encoding preserves the order.

Use Label Encoding when there is no meaningful order among categories, and they are more like nominal categories. For example, when encoding different colors or types of fruits.

**Q2. Explain how Target Guided Ordinal Encoding works and provide an example of when you might use it in a machine learning project.**

Target Guided Ordinal Encoding assigns integer labels to categorical variables based on their relationship with the target variable. It calculates the mean of the target variable for each category, orders categories by these means, and assigns labels accordingly. This method is useful when you want to capture the impact of categorical variables on the target variable, making it valuable for predictive modeling. For instance, in a customer purchase prediction task, you might apply Target Guided Ordinal Encoding to the "Product Category" feature.

**Q3. Define covariance and explain why it is important in statistical analysis. How is covariance calculated?**

**Covariance:**
Covariance is a statistical measure that quantifies the degree to which two random variables change together. It indicates whether an increase in one variable corresponds to an increase or decrease in another. It is important in statistical analysis because it helps identify relationships and dependencies between variables.

**Cov(X, Y) = Σ [(X_i - X̄)(Y_i - Ȳ)] / (n - 1)**

Where:
- X and Y are the two random variables.
- X̄ and Ȳ are the means of X and Y, respectively.
- X_i and Y_i are individual data points.
- n is the number of data points.

**Q4. For a dataset with the following categorical variables: Color (red, green, blue), Size (small, medium, large), and Material (wood, metal, plastic), perform label encoding using Python's scikit-learn library. Show your code and explain the output.**

In this code, label encoding is applied to each categorical column using scikit-learn's LabelEncoder. 

It transforms categorical values into numerical labels.

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Sample data
data = {
    'Color': ['red', 'green', 'blue', 'red', 'green'],
    'Size': ['medium', 'small', 'large', 'medium', 'small'],
    'Material': ['wood', 'metal', 'plastic', 'wood', 'plastic']
}
df = pd.DataFrame(data)
label_encoder = LabelEncoder()

# Apply Label Encoding to each column
for column in df.columns:
    df[column] = label_encoder.fit_transform(df[column])

df
```

**Explanation:** Each unique category in the categorical variables is assigned a unique integer label using Label Encoding.

**Q5. Calculate the covariance matrix for the following variables in a dataset: Age, Income, and Education level. Interpret the results.**

```python
import numpy as np

# Sample data
age = [30, 35, 40, 45, 50]
income = [50000, 60000, 75000, 80000, 90000]
education_level = [12, 16, 18, 20, 14]

# Create a data matrix
data_matrix = np.array([age, income, education_level])

# Calculate the covariance matrix
covariance_matrix = np.cov(data_matrix)

print(covariance_matrix)
data_matrix
```

**Interpretation:**
- The covariance matrix will be a 3x3 matrix since we have three variables.
- The diagonal elements of the matrix represent the variances of each variable.

**Q7. You are analyzing a dataset with two continuous variables, "Temperature" and "Humidity," and two
categorical variables, "Weather Condition" (Sunny/Cloudy/Rainy) and "Wind Direction" (North/South/
East/West). Calculate the covariance between each pair of variables and interpret the results.**


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Temperature': [25, 28, 22, 20, 30],
    'Humidity': [60, 65, 70, 75, 55],
    'Weather Condition': ['Sunny', 'Cloudy', 'Rainy', 'Sunny', 'Cloudy'],
    'Wind Direction': ['North', 'South', 'East', 'West', 'North']
}
df = pd.DataFrame(data)

# one-hot encoding for categorical variables
df_encoded = pd.get_dummies(df, columns=['Weather Condition', 'Wind Direction'])

# Calculate the covariance matrix
covariance_matrix = df_encoded.cov()

print(covariance_matrix)

plt.figure(figsize=(10, 6))
plt.imshow(covariance_matrix)
plt.colorbar(label='Covariance')
plt.xticks(np.arange(len(covariance_matrix.columns)), covariance_matrix.columns, rotation=90)
plt.yticks(np.arange(len(covariance_matrix.columns)), covariance_matrix.columns)
plt.title('Covariance Matrix Heatmap')
plt.show()
```
